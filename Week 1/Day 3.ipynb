```python
def print_list(lst):
    if isinstance(lst,list):
        for item in lst:
            print(item)
    else:
        print('Pass paramter as list')
print_list([10,20,30,40,50])
```

```python
def print_list(lst):
    sum=0
    if isinstance(lst,list):
        for item in lst:
            sum+=item
        print('Sum of',lst,':',sum)
    else:
        print('Pass paramter as list')
print_list([10,20,30,40,50])
```

```python
def print_list(lst):
    if isinstance(lst,list):
        print('Sum of',lst,':',sum(lst))
    else:
        print('Pass paramter as list')
print_list([10,20,30,40,50])
```
