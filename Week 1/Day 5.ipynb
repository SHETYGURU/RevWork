```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Year': [2019, 2020, 2021, 2022, 2023],
    'Revenue': [100, 120, 90, 150, 180],
    'Profit': [20, 30, 10, 50, 60]
}

df = pd.DataFrame(data)

df.plot(
    x='Year',
    y=['Revenue', 'Profit']
)

plt.title('Company Revenue and Profit Over Years')
plt.xlabel('Year')
plt.ylabel('Amount in $K')
plt.legend(title='Metrics')

plt.show()
```

```python
df.plot.bar(
    x='Year',
    y=['Revenue', 'Profit']
)
```

```python
df.plot.box(
    x='Year',
    y=['Revenue', 'Profit']
)
```

```python
df.plot.density(
    x='Year',
    y=['Revenue', 'Profit']
)
```

```python
data = {
    'City': ['Delhi', 'Mumbai', 'Bangalore', 'Chennai'],
    'Sales': [400, 300, 350, 200]
}
df = pd.DataFrame(data)
colors=['red','blue','green','yellow']
df.plot(x='City', y='Sales', kind='bar', color=colors, legend=False)
plt.title("Sales by City")
plt.xlabel("City")
plt.ylabel("Sales (in $k)")
plt.grid('both')
plt.show()


```

```python
colors = ['red', 'blue', 'green', 'orange']
df.plot(x='City', y='Sales', kind='barh',  legend=False,color=colors)
plt.title("Sales by City")
plt.xlabel("Sales")
plt.ylabel("City")
plt.tight_layout()
plt.show()
```

```python
from collections import Counter

c = Counter(['a', 'b', 'a', 'c', 'a'])

print(c)
```

```python
from collections import Counter

c = Counter("banana")

print(c)
```

```python
c = Counter("banana")

c.update("apple")

print(c)
```

```python
c = Counter(a=5, b=3)
c.subtract({'a': 2, 'b': 1})

print(c)
```

```python
from collections import Counter

c1 = Counter(a=3, b=2)
c2 = Counter(a=1, b=4)

print(c1 + c2)
```

```python
print(c1 & c2) #Takes minimum counts.
print(c1 | c2) #Takes maximum counts.
```

```python
from collections import OrderedDict as od

d = od()

d['a'] = 10
d['b'] = 20
d['c'] = 30

print(d)
```

```python

d = od([
    ('a', 1),
    ('b', 2),
    ('c', 3)
])

d.popitem()

print(d)
```

```python
d.popitem(last=False)

print(d)
```

```python

d1 = od([('a', 1), ('b', 2)])
d2 = od([('b', 2), ('a', 1)])

print(d1 == d2)   #False ,Because order matters in OrderedDict
```

```python

d1 = {'a': 1, 'b': 2}
d2 = {'b': 2, 'a': 1}

print(d1 == d2)  #Normal Dict
```

```python
def numbers():
    for i in range(1,6):
        yield i
```
