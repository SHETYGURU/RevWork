```python
class Car:
#Parameterized Constructor
    def __init__(self, model):
        self.model = model

car1 = Car("Camry")

print(car1.model)
```

```python
#Class Variable
class Car:
    brand = "Toyota"

car1 = Car()
car2 = Car()

print(car1.brand)
print(car2.brand)

#Instance Variable
class Car:

    def __init__(self, model):
        self.model = model

car1 = Car("Camry")
car2 = Car("Corolla")

print(car1.model)
print(car2.model)
```
